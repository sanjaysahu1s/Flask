Q1. What is Flask Framework? What are the advantages of Flask Framework?

#Answer

Flask is a lightweight web framework written in Python. It is designed to be simple, easy to use, and flexible, making it a popular choice for developing web applications and APIs.

Advantages of Flask Framework:

1. Lightweight and Easy to Get Started: Flask is a micro-framework, which means it has a small core and doesn't require any particular tools or libraries. It has a minimalistic approach and provides only the essential features needed to build web applications. This makes it easy to learn and get started with Flask.

2. Flexible and Extensible: Flask allows developers to choose and integrate different components as per their project requirements. It follows a modular design, allowing developers to add or remove features based on their needs. Flask doesn't impose any specific project structure or dependencies, giving developers the freedom to structure their application as they see fit.

3. Built-in Development Server and Debugger: Flask comes with a built-in development server and a powerful debugger, making it convenient for development and debugging purposes. The development server automatically reloads the application when changes are made, making the development process smooth and efficient.

4. Flask Extensions: Flask has a vibrant ecosystem with a wide range of extensions available. These extensions provide additional functionality and features, such as database integration, form validation, authentication, caching, and more. Flask extensions make it easy to add advanced features to your application without reinventing the wheel.

5. Well-Documented and Active Community: Flask has extensive documentation that covers various aspects of the framework, making it easy for developers to understand and use its features. Additionally, Flask has a large and active community of developers who contribute to its development, provide support, and create third-party extensions. The active community ensures that Flask stays up to date and provides solutions to common challenges.

6. Scalability: Although Flask is considered a micro-framework, it is scalable and can handle small to large-scale applications. It allows developers to scale their application by leveraging external tools and libraries as needed. Flask can be combined with other technologies like SQLAlchemy for database management or integrating with other services to handle complex requirements.

Overall, Flask provides a simple and flexible framework for building web applications in Python. Its lightweight nature, extensibility, and active community make it a popular choice for developers looking for a minimalistic yet powerful web framework.

Q3. What is App routing in Flask? Why do we use app routes?

#Answer

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or view handlers in your application. It allows you to define different routes for different parts of your application, enabling you to handle different HTTP requests and provide appropriate responses.

The app.route() decorator is used in Flask to define routes. By decorating a function with app.route(), you specify the URL path at which the function should be executed.



App routes are used in Flask for the following reasons:

URL mapping: App routes provide a way to map URLs to specific functions. This allows you to create a logical structure for your application and handle different URLs differently.

HTTP methods: Routes can be associated with specific HTTP methods such as GET, POST, PUT, DELETE, etc. This allows you to handle different types of requests and perform appropriate actions based on the HTTP method.

Dynamic routes: Flask supports dynamic routes by using variable parts in the URL path. This allows you to define routes with placeholders that can match variable values, enabling you to handle different URLs with a single route.

Modular application structure: By using app routes, you can organize your application into different views or modules, each responsible for handling specific routes. This promotes code modularity and maintainability.

RESTful APIs: For building RESTful APIs, app routes play a crucial role in defining routes for different API endpoints, allowing you to handle various CRUD (Create, Read, Update, Delete) operations.

Overall, app routing in Flask provides a flexible way to handle different URLs, HTTP methods, and build modular web applications or APIs.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
#Answer

from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    '''

if __name__ == '__main__':
    app.run()
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

#Answer

In Flask, the url_for() function is used for URL building. It is a helpful function that generates a URL for a given view function. It takes the name of the view function as an argument and returns the URL associated with that function.

Here's an example Python code to demonstrate the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'Contact us at contact@example.com'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is the profile page for {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))                 
        print(url_for('about'))                 
        print(url_for('contact'))              
        print(url_for('user_profile', username='john'))
```
